```python
# Fixed Plotting Functions for Jupyter Notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import warnings

# CLEAN MATPLOTLIB SETUP FOR JUPYTER
def setup_plotting_environment():
    """Clean plotting setup for Jupyter notebooks"""
    # Clear any existing configuration
    plt.rcdefaults()
    
    # Use inline backend for Jupyter
    %matplotlib inline
    
    # Set basic style
    plt.style.use('default')
    sns.set_style("whitegrid")
    sns.set_palette("husl")
    
    # Configure matplotlib for better display
    plt.rcParams.update({
        'figure.figsize': (12, 8),
        'figure.dpi': 100,
        'savefig.dpi': 300,
        'font.size': 11,
        'axes.titlesize': 14,
        'axes.labelsize': 12,
        'xtick.labelsize': 10,
        'ytick.labelsize': 10,
        'legend.fontsize': 10,
        'axes.grid': True,
        'grid.alpha': 0.3,
        'figure.facecolor': 'white',
        'axes.facecolor': 'white'
    })
    
    # Suppress warnings
    warnings.filterwarnings('ignore')
    print("‚úÖ Plotting environment configured for Jupyter!")

def test_plotting():
    """Test plotting with explicit figure creation"""
    # Create new figure explicitly
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111)
    
    # Create test data
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    
    # Plot
    ax.plot(x, y, label='sin(x)', linewidth=2, color='blue')
    ax.set_title('‚úÖ Test Plot - Matplotlib is Working!', fontsize=14, pad=20)
    ax.set_xlabel('X Values')
    ax.set_ylabel('Y Values')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Force display
    plt.tight_layout()
    plt.show()
    
    # Close figure to free memory
    plt.close(fig)
    
    print("‚úÖ If you see the plot above, matplotlib is working!")
    return True

def plot_sentiment_distribution(df, save_path=None):
    """Fixed sentiment distribution with explicit figure management"""
    print("üìä Creating sentiment distribution plots...")
    
    # Find available sentiment columns
    sentiment_columns = []
    column_labels = {}
    
    possible_columns = {
        'sentiment_overall_enhanced': 'Enhanced Overall',
        'sentiment_base_textblob': 'TextBlob Base',
        'sentiment_lexicon_normalized': 'Lexicon Normalized',
        'sentiment_proximity_enhanced': 'Proximity Enhanced',
        'sentiment_recency_weighted': 'Recency Weighted',
        'sentiment_negation_adjusted': 'Negation Adjusted'
    }
    
    for col, label in possible_columns.items():
        if col in df.columns and not df[col].isna().all():
            sentiment_columns.append(col)
            column_labels[col] = label
    
    if len(sentiment_columns) == 0:
        print("‚ùå No valid sentiment columns found!")
        return
    
    print(f"üìà Found {len(sentiment_columns)} sentiment columns")
    
    # Calculate grid layout
    n_plots = len(sentiment_columns)
    n_cols = min(3, n_plots)
    n_rows = (n_plots + n_cols - 1) // n_cols
    
    # Create figure explicitly
    fig = plt.figure(figsize=(6*n_cols, 5*n_rows))
    fig.suptitle('Sentiment Analysis Distribution', fontsize=16, y=0.98)
    
    colors = sns.color_palette("husl", n_plots)
    
    for i, col in enumerate(sentiment_columns):
        # Create subplot explicitly
        ax = fig.add_subplot(n_rows, n_cols, i+1)
        
        # Clean data
        data = df[col].dropna()
        if len(data) == 0:
            ax.text(0.5, 0.5, 'No Data', ha='center', va='center', transform=ax.transAxes)
            ax.set_title(column_labels[col])
            continue
        
        # Create histogram
        ax.hist(data, bins=50, alpha=0.7, color=colors[i], edgecolor='black', linewidth=0.5)
        
        # Add statistics
        mean_val = data.mean()
        median_val = data.median()
        std_val = data.std()
        
        ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, 
                  label=f'Mean: {mean_val:.3f}')
        ax.axvline(median_val, color='orange', linestyle=':', linewidth=2,
                  label=f'Median: {median_val:.3f}')
        
        ax.set_title(column_labels[col], fontsize=12, pad=10)
        ax.set_xlabel('Sentiment Score')
        ax.set_ylabel('Frequency')
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3)
        
        # Add text box with stats
        stats_text = f'n={len(data):,}\\nstd={std_val:.3f}'
        ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, 
                verticalalignment='top', fontsize=8,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Adjust layout and display
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)  # Close to free memory
    print("‚úÖ Sentiment distribution plot complete!")

def plot_industry_sentiment(df, save_path=None):
    """Fixed industry sentiment with explicit figure management"""
    print("üìä Creating industry sentiment plots...")
    
    if 'primary_industry' not in df.columns:
        print("‚ùå No 'primary_industry' column found!")
        return
    
    # Find best sentiment column
    sentiment_col = None
    for col in ['sentiment_overall_enhanced', 'sentiment_base_textblob']:
        if col in df.columns and not df[col].isna().all():
            sentiment_col = col
            break
    
    if sentiment_col is None:
        print("‚ùå No valid sentiment column found!")
        return
    
    print(f"üìà Using sentiment column: {sentiment_col}")
    
    # Filter and prepare data
    df_filtered = df.dropna(subset=['primary_industry', sentiment_col])
    print(f"üìã Data after filtering: {len(df_filtered):,} articles")
    
    if len(df_filtered) == 0:
        print("‚ùå No data remaining after filtering!")
        return
    
    # Get top industries with minimum article count
    industry_counts = df_filtered['primary_industry'].value_counts()
    top_industries = industry_counts[industry_counts >= 50].head(8).index
    df_plot = df_filtered[df_filtered['primary_industry'].isin(top_industries)]
    
    if len(df_plot) == 0:
        print("‚ùå No industries with sufficient data!")
        return
    
    # Create figure explicitly
    fig = plt.figure(figsize=(16, 8))
    
    # Create subplots
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    # Box plot
    try:
        # Prepare data for box plot
        box_data = []
        box_labels = []
        for industry in top_industries:
            industry_data = df_plot[df_plot['primary_industry'] == industry][sentiment_col]
            if len(industry_data) > 0:
                box_data.append(industry_data)
                box_labels.append(industry.replace('_', ' ').title())
        
        # Create box plot
        bp = ax1.boxplot(box_data, labels=box_labels, patch_artist=True)
        
        # Color the boxes
        colors = sns.color_palette("husl", len(box_data))
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)
        
        ax1.set_title('Sentiment Distribution by Industry', fontsize=14, pad=20)
        ax1.set_xlabel('Industry', fontsize=12)
        ax1.set_ylabel('Sentiment Score', fontsize=12)
        ax1.tick_params(axis='x', rotation=45)
        ax1.grid(True, alpha=0.3)
        
    except Exception as e:
        print(f"‚ö†Ô∏è Box plot error: {e}")
        ax1.text(0.5, 0.5, 'Box plot failed', ha='center', va='center', transform=ax1.transAxes)
    
    # Bar plot of means
    try:
        industry_sentiment = df_plot.groupby('primary_industry')[sentiment_col].agg(['mean', 'count', 'std']).reset_index()
        industry_sentiment = industry_sentiment.sort_values('mean')
        
        colors = ['red' if x < 0 else 'green' for x in industry_sentiment['mean']]
        bars = ax2.bar(range(len(industry_sentiment)), industry_sentiment['mean'], 
                      color=colors, alpha=0.7, edgecolor='black', linewidth=0.5)
        
        ax2.set_title('Average Sentiment by Industry', fontsize=14, pad=20)
        ax2.set_xlabel('Industry', fontsize=12)
        ax2.set_ylabel('Average Sentiment Score', fontsize=12)
        ax2.axhline(0, color='black', linestyle='-', alpha=0.5, linewidth=1)
        
        # Set x-axis labels
        industry_labels = [ind.replace('_', ' ').title() for ind in industry_sentiment['primary_industry']]
        ax2.set_xticks(range(len(industry_sentiment)))
        ax2.set_xticklabels(industry_labels, rotation=45, ha='right')
        
        # Add value labels on bars
        for i, (bar, value, count) in enumerate(zip(bars, industry_sentiment['mean'], industry_sentiment['count'])):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., 
                    height + 0.01 if height >= 0 else height - 0.02,
                    f'{value:.3f}\\n(n={count:,})', 
                    ha='center', va='bottom' if height >= 0 else 'top',
                    fontsize=8)
        
        ax2.grid(True, alpha=0.3)
        
    except Exception as e:
        print(f"‚ö†Ô∏è Bar plot error: {e}")
        ax2.text(0.5, 0.5, 'Bar plot failed', ha='center', va='center', transform=ax2.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Industry sentiment plot complete!")

def plot_job_sentiment(df, save_path=None):
    """Fixed job sentiment with explicit figure management"""
    print("üìä Creating job sentiment plot...")
    
    if 'primary_job' not in df.columns:
        print("‚ùå No 'primary_job' column found!")
        return
    
    # Find sentiment column
    sentiment_col = None
    for col in ['sentiment_overall_enhanced', 'sentiment_base_textblob']:
        if col in df.columns and not df[col].isna().all():
            sentiment_col = col
            break
    
    if sentiment_col is None:
        print("‚ùå No valid sentiment column found!")
        return
    
    # Filter and prepare data
    df_filtered = df.dropna(subset=['primary_job', sentiment_col])
    
    if len(df_filtered) == 0:
        print("‚ùå No data after filtering!")
        return
    
    job_sentiment = df_filtered.groupby('primary_job')[sentiment_col].agg(['mean', 'count', 'std']).reset_index()
    job_sentiment = job_sentiment[job_sentiment['count'] >= 20]  # Minimum count
    job_sentiment = job_sentiment.sort_values('mean')
    
    if len(job_sentiment) == 0:
        print("‚ùå No job categories with sufficient data!")
        return
    
    print(f"üìà Job categories with data: {len(job_sentiment)}")
    
    # Create figure explicitly
    fig = plt.figure(figsize=(12, max(8, len(job_sentiment) * 0.5)))
    ax = fig.add_subplot(111)
    
    # Create horizontal bar plot
    colors = ['darkred' if x < -0.05 else 'red' if x < 0 else 'lightgreen' if x < 0.05 else 'green' 
              for x in job_sentiment['mean']]
    
    y_pos = range(len(job_sentiment))
    bars = ax.barh(y_pos, job_sentiment['mean'], color=colors, alpha=0.7, 
                   edgecolor='black', linewidth=0.5)
    
    ax.set_title('Average AI Sentiment by Job Category', fontsize=14, pad=20)
    ax.set_xlabel('Average Sentiment Score', fontsize=12)
    ax.set_ylabel('Job Category', fontsize=12)
    ax.axvline(0, color='black', linestyle='-', alpha=0.5, linewidth=1)
    
    # Set y-axis labels
    job_labels = [job.replace('_', ' ').title() for job in job_sentiment['primary_job']]
    ax.set_yticks(y_pos)
    ax.set_yticklabels(job_labels)
    
    # Add value labels
    for i, (value, count) in enumerate(zip(job_sentiment['mean'], job_sentiment['count'])):
        x_pos = value + 0.01 if value >= 0 else value - 0.01
        ax.text(x_pos, i, f'{value:.3f}\\n(n={count:,})', 
                ha='left' if value >= 0 else 'right', va='center', fontsize=9)
    
    ax.grid(True, alpha=0.3, axis='x')
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Job sentiment plot complete!")

def plot_technology_mentions(df, save_path=None):
    """Fixed technology mentions with explicit figure management"""
    print("üìä Creating technology mentions plot...")
    
    if 'ai_technologies' not in df.columns:
        print("‚ùå No 'ai_technologies' column found!")
        return
    
    # Count technology mentions
    tech_counts = defaultdict(int)
    specific_tech_counts = defaultdict(int)
    
    for tech_dict in df['ai_technologies'].dropna():
        if isinstance(tech_dict, dict):
            for category, technologies in tech_dict.items():
                if category == 'specific_models':
                    if isinstance(technologies, list):
                        for tech in technologies:
                            specific_tech_counts[tech] += 1
                else:
                    if isinstance(technologies, list):
                        tech_counts[category] += len(technologies)
                    else:
                        tech_counts[category] += 1
    
    if not tech_counts and not specific_tech_counts:
        print("‚ùå No technology mentions found!")
        return
    
    # Create figure explicitly
    fig = plt.figure(figsize=(16, 8))
    
    # Create subplots
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    # Plot 1: Technology categories
    if tech_counts:
        categories = list(tech_counts.keys())
        counts = list(tech_counts.values())
        
        # Clean category names
        clean_categories = [cat.replace('_', ' ').title() for cat in categories]
        
        bars1 = ax1.bar(range(len(categories)), counts, 
                       color=sns.color_palette("viridis", len(categories)),
                       edgecolor='black', linewidth=0.5)
        
        ax1.set_title('AI Technology Categories Mentioned', fontsize=14, pad=20)
        ax1.set_xlabel('Technology Category', fontsize=12)
        ax1.set_ylabel('Number of Mentions', fontsize=12)
        ax1.set_xticks(range(len(categories)))
        ax1.set_xticklabels(clean_categories, rotation=45, ha='right')
        
        # Add value labels
        for bar, value in zip(bars1, counts):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + max(counts)*0.01,
                    f'{value:,}', ha='center', va='bottom', fontsize=10)
        
        ax1.grid(True, alpha=0.3, axis='y')
    else:
        ax1.text(0.5, 0.5, 'No technology categories found', 
                ha='center', va='center', transform=ax1.transAxes)
        ax1.set_title('Technology Categories')
    
    # Plot 2: Specific AI models (top 10)
    if specific_tech_counts:
        sorted_models = sorted(specific_tech_counts.items(), key=lambda x: x[1], reverse=True)[:10]
        models, model_counts = zip(*sorted_models)
        
        bars2 = ax2.bar(range(len(models)), model_counts,
                       color=sns.color_palette("plasma", len(models)),
                       edgecolor='black', linewidth=0.5)
        
        ax2.set_title('Top AI Models Mentioned', fontsize=14, pad=20)
        ax2.set_xlabel('AI Model', fontsize=12)
        ax2.set_ylabel('Number of Mentions', fontsize=12)
        ax2.set_xticks(range(len(models)))
        ax2.set_xticklabels([m.upper() for m in models], rotation=45, ha='right')
        
        # Add value labels
        for bar, value in zip(bars2, model_counts):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + max(model_counts)*0.01,
                    f'{value:,}', ha='center', va='bottom', fontsize=10)
        
        ax2.grid(True, alpha=0.3, axis='y')
    else:
        ax2.text(0.5, 0.5, 'No specific AI models found', 
                ha='center', va='center', transform=ax2.transAxes)
        ax2.set_title('Specific AI Models')
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Technology mentions plot complete!")

def plot_sentiment_comparison(df, save_path=None):
    """Fixed sentiment comparison with explicit figure management"""
    print("üìä Creating sentiment comparison plots...")
    
    # Find available sentiment columns
    sentiment_cols = []
    labels = []
    
    column_mapping = {
        'sentiment_overall_enhanced': 'Enhanced Overall',
        'sentiment_base_textblob': 'TextBlob',
        'sentiment_lexicon_normalized': 'Lexicon (Normalized)',
        'sentiment_proximity_enhanced': 'Proximity Enhanced',
        'sentiment_recency_weighted': 'Recency Weighted',
        'sentiment_negation_adjusted': 'Negation Adjusted'
    }
    
    for col, label in column_mapping.items():
        if col in df.columns and not df[col].isna().all():
            sentiment_cols.append(col)
            labels.append(label)
    
    if len(sentiment_cols) < 2:
        print("‚ùå Need at least 2 sentiment measures to compare!")
        return
    
    print(f"üìà Comparing {len(sentiment_cols)} sentiment methods")
    
    # Create figure explicitly
    fig = plt.figure(figsize=(16, 8))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    # Correlation heatmap
    try:
        corr_data = df[sentiment_cols].corr()
        
        mask = np.triu(np.ones_like(corr_data, dtype=bool))
        sns.heatmap(corr_data, annot=True, cmap='RdBu_r', center=0,
                   square=True, mask=mask,
                   xticklabels=labels, yticklabels=labels, 
                   ax=ax1, cbar_kws={'shrink': 0.8})
        ax1.set_title('Sentiment Methods Correlation', fontsize=14, pad=20)
        ax1.tick_params(axis='x', rotation=45)
        ax1.tick_params(axis='y', rotation=0)
        
    except Exception as e:
        print(f"‚ö†Ô∏è Correlation plot error: {e}")
        ax1.text(0.5, 0.5, 'Correlation plot failed', 
                ha='center', va='center', transform=ax1.transAxes)
    
    # Box plot comparison
    try:
        # Prepare data for melting
        df_plot = df[sentiment_cols].copy()
        df_plot.columns = labels
        
        df_melted = df_plot.melt(var_name='Sentiment_Method', value_name='Score')
        df_melted = df_melted.dropna()
        
        # Create box plot manually for better control
        method_data = []
        method_labels = []
        for method in labels:
            method_scores = df_melted[df_melted['Sentiment_Method'] == method]['Score']
            if len(method_scores) > 0:
                method_data.append(method_scores)
                method_labels.append(method)
        
        bp = ax2.boxplot(method_data, labels=method_labels, patch_artist=True)
        
        # Color the boxes
        colors = sns.color_palette("husl", len(method_data))
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)
        
        ax2.set_title('Sentiment Score Distributions', fontsize=14, pad=20)
        ax2.set_xlabel('Sentiment Method', fontsize=12)
        ax2.set_ylabel('Sentiment Score', fontsize=12)
        ax2.tick_params(axis='x', rotation=45)
        ax2.axhline(0, color='black', linestyle='--', alpha=0.5)
        ax2.grid(True, alpha=0.3)
        
        # Add median values as text
        for i, data in enumerate(method_data):
            median_val = np.median(data)
            ax2.text(i+1, median_val + 0.02, f'{median_val:.3f}', 
                    ha='center', va='bottom', fontweight='bold', fontsize=9)
        
    except Exception as e:
        print(f"‚ö†Ô∏è Box plot error: {e}")
        ax2.text(0.5, 0.5, 'Box plot failed', 
                ha='center', va='center', transform=ax2.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Sentiment comparison plot complete!")

def create_fixed_sentiment_dashboard(df):
    """Create dashboard with fixed plotting - each plot manages its own figure"""
    print("üéØ Creating FIXED comprehensive sentiment analysis dashboard...")
    print("=" * 70)
    
    # Setup plotting environment first
    setup_plotting_environment()
    
    # Test plotting
    print("üß™ Testing plotting configuration...")
    if not test_plotting():
        print("‚ùå Plotting test failed! Please check your environment.")
        return False
    
    print(f"üìä Dataset info: {df.shape[0]:,} articles, {df.shape[1]} columns")
    
    # Check available columns
    available_cols = [col for col in df.columns if 'sentiment' in col.lower() or 
                     col in ['primary_industry', 'primary_job', 'ai_technologies']]
    print(f"üìã Available analysis columns: {len(available_cols)}")
    
    # Create all plots with individual figure management
    plots = [
        ("Sentiment Distribution", plot_sentiment_distribution),
        ("Sentiment by Industry", plot_industry_sentiment), 
        ("Sentiment by Job Category", plot_job_sentiment),
        ("AI Technology Mentions", plot_technology_mentions),
        ("Sentiment Methods Comparison", plot_sentiment_comparison)
    ]
    
    success_count = 0
    for i, (title, plot_func) in enumerate(plots, 1):
        print(f"\\n{i}. {title}:")
        print("-" * 50)
        try:
            plot_func(df)
            success_count += 1
        except Exception as e:
            print(f"‚ùå Error in {title}: {e}")
            import traceback
            traceback.print_exc()
    
    print(f"\\nüéâ Dashboard complete! {success_count}/{len(plots)} plots successful")
    return success_count == len(plots)

# MAIN EXECUTION FUNCTION
def run_fixed_plotting_pipeline():
    """Run the complete pipeline with fixed plotting"""
    print("üöÄ Starting FIXED plotting pipeline...")
    print("=" * 80)
    
    # Load your cached data
    try:
        # Assuming you have the load_from_cache function from your original code
        df_enhanced = load_from_cache('fast_enhanced_data_with_features.pkl')
        
        if df_enhanced is None:
            print("‚ùå No cached data found. Please run the sentiment analysis pipeline first.")
            return None
        
        print(f"‚úÖ Loaded cached data with {len(df_enhanced):,} articles")
        
        # Create the fixed dashboard
        success = create_fixed_sentiment_dashboard(df_enhanced)
        
        if success:
            print("\\nüéØ All plots created successfully!")
        else:
            print("\\n‚ö†Ô∏è Some plots failed. Check the error messages above.")
            
        return df_enhanced
        
    except Exception as e:
        print(f"‚ùå Pipeline error: {e}")
        import traceback
        traceback.print_exc()
        return None

def plot_ai_technology_adoption_over_time(df, save_path=None):
    """Plot AI technology adoption trends over time using actual data"""
    print("üìä Creating AI technology adoption over time plot...")
    
    # Check if we have date information
    date_col = None
    for col in ['date', 'published_date', 'publish_date', 'timestamp']:
        if col in df.columns:
            date_col = col
            break
    
    if date_col is None:
        print("‚ùå No date column found for time series analysis!")
        return
    
    # Check if we have ai_technologies column with actual data
    if 'ai_technologies' not in df.columns:
        print("‚ùå No 'ai_technologies' column found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(14, 8))
    ax = fig.add_subplot(111)
    
    try:
        # Convert date column to datetime
        df_copy = df.copy()
        df_copy[date_col] = pd.to_datetime(df_copy[date_col], errors='coerce')
        df_copy = df_copy.dropna(subset=[date_col, 'ai_technologies'])
        
        if len(df_copy) == 0:
            print("‚ùå No valid dates or AI technology data found!")
            return
        
        # Extract actual technologies from the data
        all_technologies = set()
        for tech_dict in df_copy['ai_technologies'].dropna():
            if isinstance(tech_dict, dict):
                for category, tech_list in tech_dict.items():
                    if isinstance(tech_list, list):
                        all_technologies.update(tech_list[:3])  # Limit to avoid overcrowding
                    else:
                        all_technologies.add(category)
        
        if not all_technologies:
            print("‚ùå No technology data found in ai_technologies column!")
            return
        
        # Take top 5 most frequent technologies
        tech_frequency = defaultdict(int)
        for tech_dict in df_copy['ai_technologies'].dropna():
            if isinstance(tech_dict, dict):
                for category, tech_list in tech_dict.items():
                    if isinstance(tech_list, list):
                        for tech in tech_list:
                            tech_frequency[tech] += 1
                    else:
                        tech_frequency[category] += 1
        
        top_technologies = [tech for tech, _ in sorted(tech_frequency.items(), key=lambda x: x[1], reverse=True)[:5]]
        
        # Group by month-year for better visualization
        df_copy['year_month'] = df_copy[date_col].dt.to_period('M')
        
        # Count actual technology mentions by time period
        tech_counts = {}
        
        for tech_name in top_technologies:
            monthly_counts = []
            periods = []
            
            for period in sorted(df_copy['year_month'].unique()):
                period_data = df_copy[df_copy['year_month'] == period]
                count = 0
                
                for tech_dict in period_data['ai_technologies'].dropna():
                    if isinstance(tech_dict, dict):
                        for category, tech_list in tech_dict.items():
                            if isinstance(tech_list, list):
                                count += tech_list.count(tech_name)
                            elif category == tech_name:
                                count += 1
                
                monthly_counts.append(count)
                periods.append(period)
            
            tech_counts[tech_name] = {
                'periods': [str(p) for p in periods],
                'counts': monthly_counts
            }
        
        # Plot each technology
        colors = plt.cm.Set1(np.linspace(0, 1, len(top_technologies)))
        
        for i, (tech_name, data) in enumerate(tech_counts.items()):
            if data['counts'] and max(data['counts']) > 0:
                ax.plot(range(len(data['periods'])), data['counts'], 
                       marker='o', linewidth=2, markersize=6,
                       label=tech_name.replace('_', ' ').title(), color=colors[i])
        
        # Format plot
        ax.set_title('AI Technology Adoption Over Time', fontsize=16, pad=20)
        ax.set_xlabel('Date', fontsize=12)
        ax.set_ylabel('Mentions', fontsize=12)
        ax.legend(title='Technology', bbox_to_anchor=(1.05, 1), loc='upper left')
        ax.grid(True, alpha=0.3)
        
        # Set x-axis labels
        if tech_counts:
            first_tech_data = list(tech_counts.values())[0]
            if first_tech_data['periods']:
                x_labels = first_tech_data['periods']
                step = max(1, len(x_labels)//8)
                ax.set_xticks(range(0, len(x_labels), step))
                ax.set_xticklabels([x_labels[i] for i in range(0, len(x_labels), step)], 
                                  rotation=45, ha='right')
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating time series plot: {e}")
        ax.text(0.5, 0.5, f'Time series plot failed:\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ AI technology adoption over time plot complete!")

def plot_topic_correlation_heatmap(df, save_path=None):
    """Plot correlation heatmap between topics using actual topic data only"""
    print("üìä Creating topic correlation heatmap...")
    
    # Look for actual topic-related columns
    topic_cols = [col for col in df.columns if 'topic' in col.lower() and ('probability' in col.lower() or 'score' in col.lower())]
    
    if not topic_cols:
        # Try alternative patterns
        topic_cols = [col for col in df.columns if col.startswith('topic_') or col.endswith('_topic')]
    
    if not topic_cols:
        print("‚ùå No topic probability/score columns found!")
        print("üí° Expected columns like 'topic_0_probability', 'topic_score', etc.")
        return
    
    # Create figure
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    
    try:
        topic_df = df[topic_cols].dropna()
        
        if topic_df.empty:
            print("‚ùå No valid topic data available for correlation analysis!")
            return
        
        # Calculate correlation matrix
        correlation_matrix = topic_df.corr()
        
        # Create heatmap
        sns.heatmap(correlation_matrix, 
                   annot=True, 
                   cmap='RdBu_r', 
                   center=0,
                   square=True,
                   fmt='.2f',
                   cbar_kws={'shrink': 0.8},
                   ax=ax)
        
        ax.set_title('Topic Correlation Heatmap', fontsize=16, pad=20)
        ax.tick_params(axis='x', rotation=45, labelsize=10)
        ax.tick_params(axis='y', rotation=0, labelsize=10)
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating correlation heatmap: {e}")
        ax.text(0.5, 0.5, f'Correlation heatmap failed:\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Topic correlation heatmap complete!")

def plot_ai_job_network(df, save_path=None):
    """Plot improved network graph showing relationships between AI technologies and job categories"""
    print("üìä Creating improved AI technology and job categories network...")
    
    try:
        import networkx as nx
    except ImportError:
        print("‚ùå NetworkX not installed. Install with: pip install networkx")
        return
    
    if 'ai_technologies' not in df.columns or 'primary_job' not in df.columns:
        print("‚ùå Required columns 'ai_technologies' and 'primary_job' not found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(16, 12))
    ax = fig.add_subplot(111)
    
    try:
        # Create network graph
        G = nx.Graph()
        
        # Count actual co-occurrences between technologies and jobs
        tech_job_counts = defaultdict(int)
        tech_sentiment = defaultdict(list)
        job_sentiment = defaultdict(list)
        
        # Get sentiment column
        sentiment_col = 'sentiment_overall_enhanced' if 'sentiment_overall_enhanced' in df.columns else 'sentiment_base_textblob'
        
        for idx, row in df.iterrows():
            if pd.isna(row['primary_job']) or pd.isna(row['ai_technologies']):
                continue
                
            job = row['primary_job']
            sentiment = row.get(sentiment_col, 0) if pd.notna(row.get(sentiment_col)) else 0
            
            # Extract actual technologies from data
            ai_techs = row['ai_technologies']
            if isinstance(ai_techs, dict):
                for tech_category, tech_list in ai_techs.items():
                    if isinstance(tech_list, list):
                        for tech in tech_list[:1]:  # Limit to 1 to reduce clutter
                            tech_name = tech.replace('_', ' ')
                            tech_job_counts[(tech_name, job)] += 1
                            tech_sentiment[tech_name].append(sentiment)
                            job_sentiment[job].append(sentiment)
                    else:
                        tech_name = tech_category.replace('_', ' ')
                        tech_job_counts[(tech_name, job)] += 1
                        tech_sentiment[tech_name].append(sentiment)
                        job_sentiment[job].append(sentiment)
        
        # Filter for significant relationships and limit to top connections
        min_cooccurrence = max(10, len(df) // 5000)  # Higher threshold for cleaner network
        significant_pairs = {k: v for k, v in tech_job_counts.items() if v >= min_cooccurrence}
        
        # Keep only top 30 strongest relationships
        significant_pairs = dict(sorted(significant_pairs.items(), key=lambda x: x[1], reverse=True)[:30])
        
        if not significant_pairs:
            print(f"‚ùå No significant technology-job relationships found (threshold: {min_cooccurrence})!")
            return
        
        # Add nodes and edges based on actual data
        for (tech, job), weight in significant_pairs.items():
            # Calculate average sentiment for coloring
            tech_avg_sentiment = np.mean(tech_sentiment[tech]) if tech_sentiment[tech] else 0
            job_avg_sentiment = np.mean(job_sentiment[job]) if job_sentiment[job] else 0
            
            G.add_node(tech, node_type='technology', sentiment=tech_avg_sentiment)
            G.add_node(job, node_type='job', sentiment=job_avg_sentiment)
            G.add_edge(tech, job, weight=weight)
        
        # Use better layout algorithm
        pos = nx.spring_layout(G, k=2, iterations=100, seed=42)
        
        # Separate nodes by type
        tech_nodes = [(node, data) for node, data in G.nodes(data=True) if data.get('node_type') == 'technology']
        job_nodes = [(node, data) for node, data in G.nodes(data=True) if data.get('node_type') == 'job']
        
        # Draw edges with varying thickness based on weight
        edges = G.edges(data=True)
        weights = [edge[2]['weight'] for edge in edges]
        max_weight = max(weights) if weights else 1
        
        nx.draw_networkx_edges(G, pos, 
                             width=[w/max_weight * 3 for w in weights],
                             alpha=0.4, edge_color='gray', ax=ax)
        
        # Draw technology nodes with sentiment-based coloring
        tech_node_names = [node for node, _ in tech_nodes]
        tech_sentiments = [data['sentiment'] for _, data in tech_nodes]
        tech_colors = plt.cm.RdYlBu([0.5 + s*0.5 for s in tech_sentiments])  # Blue=positive, Red=negative
        
        nx.draw_networkx_nodes(G, pos, nodelist=tech_node_names,
                             node_color=tech_colors, node_size=1200,
                             alpha=0.8, label='AI Technology', ax=ax)
        
        # Draw job nodes with sentiment-based coloring
        job_node_names = [node for node, _ in job_nodes]
        job_sentiments = [data['sentiment'] for _, data in job_nodes]
        job_colors = plt.cm.RdYlGn([0.5 + s*0.5 for s in job_sentiments])  # Green=positive, Red=negative
        
        nx.draw_networkx_nodes(G, pos, nodelist=job_node_names,
                             node_color=job_colors, node_size=1000,
                             alpha=0.8, label='Job Category', node_shape='s', ax=ax)
        
        # Draw labels with better positioning
        labels = {}
        for node in G.nodes():
            if len(node) > 15:  # Truncate long labels
                labels[node] = node[:12] + "..."
            else:
                labels[node] = node
        
        nx.draw_networkx_labels(G, pos, labels, font_size=8, font_weight='bold', ax=ax)
        
        ax.set_title('AI Technology Impact on Job Categories Network\\n(Node color: Blue/Green=Enhancement, Red=Risk)', 
                    fontsize=16, pad=20)
        ax.legend(scatterpoints=1, loc='upper right')
        ax.axis('off')
        
        # Add colorbar legend
        sm1 = plt.cm.ScalarMappable(cmap=plt.cm.RdYlBu, norm=plt.Normalize(vmin=-0.5, vmax=0.5))
        sm1.set_array([])
        cbar1 = plt.colorbar(sm1, ax=ax, orientation='horizontal', pad=0.02, shrink=0.3, aspect=20)
        cbar1.set_label('Technology Sentiment (Blue=Enhancement, Red=Risk)', fontsize=10)
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating network plot: {e}")
        ax.text(0.5, 0.5, f'Network plot failed:\\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Improved AI-Job network plot complete!")

def plot_top_technologies_posing_risk(df, save_path=None):
    """Plot top AI technologies that pose the most risk to jobs based on sentiment"""
    print("üìä Creating top technologies posing risk to jobs plot...")
    
    if 'ai_technologies' not in df.columns:
        print("‚ùå No 'ai_technologies' column found!")
        return
    
    # Get sentiment column
    sentiment_col = 'sentiment_overall_enhanced' if 'sentiment_overall_enhanced' in df.columns else 'sentiment_base_textblob'
    if sentiment_col not in df.columns:
        print("‚ùå No sentiment column found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111)
    
    try:
        # Extract technology sentiment data
        tech_sentiments = defaultdict(list)
        tech_counts = defaultdict(int)
        
        for idx, row in df.iterrows():
            if pd.isna(row['ai_technologies']) or pd.isna(row[sentiment_col]):
                continue
                
            sentiment = row[sentiment_col]
            ai_techs = row['ai_technologies']
            
            if isinstance(ai_techs, dict):
                for tech_category, tech_list in ai_techs.items():
                    if isinstance(tech_list, list):
                        for tech in tech_list:
                            tech_sentiments[tech].append(sentiment)
                            tech_counts[tech] += 1
                    else:
                        tech_sentiments[tech_category].append(sentiment)
                        tech_counts[tech_category] += 1
        
        # Calculate average sentiment and filter for technologies with negative sentiment
        tech_risk_scores = {}
        min_mentions = max(10, len(df) // 1000)  # Minimum mentions threshold
        
        for tech, sentiments in tech_sentiments.items():
            if tech_counts[tech] >= min_mentions:
                avg_sentiment = np.mean(sentiments)
                if avg_sentiment < 0:  # Only negative sentiment (risk)
                    # Risk score = how negative * how many mentions
                    risk_score = abs(avg_sentiment) * np.log(tech_counts[tech] + 1)
                    tech_risk_scores[tech] = {
                        'avg_sentiment': avg_sentiment,
                        'count': tech_counts[tech],
                        'risk_score': risk_score
                    }
        
        if not tech_risk_scores:
            print("‚ùå No technologies with negative sentiment found!")
            return
        
        # Sort by risk score and take top 15
        sorted_techs = sorted(tech_risk_scores.items(), 
                            key=lambda x: x[1]['risk_score'], reverse=True)[:15]
        
        tech_names = [tech.replace('_', ' ').title() for tech, _ in sorted_techs]
        risk_scores = [data['risk_score'] for _, data in sorted_techs]
        sentiments = [data['avg_sentiment'] for _, data in sorted_techs]
        counts = [data['count'] for _, data in sorted_techs]
        
        # Create horizontal bar plot
        y_pos = range(len(tech_names))
        bars = ax.barh(y_pos, risk_scores, 
                      color='darkred', alpha=0.7, edgecolor='black', linewidth=0.5)
        
        ax.set_title('Top AI Technologies Posing Risk to Jobs\\n(Based on Negative Sentiment & Mention Frequency)', 
                    fontsize=14, pad=20)
        ax.set_xlabel('Risk Score (|Negative Sentiment| √ó log(Mentions))', fontsize=12)
        ax.set_ylabel('AI Technology', fontsize=12)
        ax.set_yticks(y_pos)
        ax.set_yticklabels(tech_names)
        
        # Add value labels
        for i, (bar, sentiment, count) in enumerate(zip(bars, sentiments, counts)):
            width = bar.get_width()
            ax.text(width + max(risk_scores) * 0.01, bar.get_y() + bar.get_height()/2,
                   f'{sentiment:.3f}\\n({count:,} mentions)', 
                   ha='left', va='center', fontsize=8)
        
        ax.invert_yaxis()
        ax.grid(True, alpha=0.3, axis='x')
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating risk technologies plot: {e}")
        ax.text(0.5, 0.5, f'Risk technologies plot failed:\\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Top technologies posing risk plot complete!")

def plot_industries_most_jobs_at_risk(df, save_path=None):
    """Plot industries with the most jobs at risk based on sentiment"""
    print("üìä Creating industries with most jobs at risk plot...")
    
    if 'primary_industry' not in df.columns:
        print("‚ùå No 'primary_industry' column found!")
        return
    
    # Get sentiment column
    sentiment_col = 'sentiment_overall_enhanced' if 'sentiment_overall_enhanced' in df.columns else 'sentiment_base_textblob'
    if sentiment_col not in df.columns:
        print("‚ùå No sentiment column found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111)
    
    try:
        # Filter data
        df_filtered = df.dropna(subset=['primary_industry', sentiment_col])
        
        if len(df_filtered) == 0:
            print("‚ùå No valid industry and sentiment data!")
            return
        
        # Calculate industry risk metrics
        industry_stats = df_filtered.groupby('primary_industry')[sentiment_col].agg([
            'mean', 'count', 'std', 
            lambda x: (x < 0).sum(),  # Count of negative sentiment
            lambda x: (x < -0.1).sum()  # Count of strongly negative sentiment
        ]).round(3)
        
        industry_stats.columns = ['avg_sentiment', 'total_articles', 'sentiment_std', 
                                'negative_count', 'strongly_negative_count']
        
        # Calculate risk score: (negative sentiment intensity) * (proportion of negative articles) * log(total articles)
        industry_stats['negative_proportion'] = industry_stats['negative_count'] / industry_stats['total_articles']
        industry_stats['risk_score'] = (
            abs(industry_stats['avg_sentiment'].clip(upper=0)) * 
            industry_stats['negative_proportion'] * 
            np.log(industry_stats['total_articles'] + 1)
        )
        
        # Filter for industries with sufficient data and some risk
        min_articles = max(50, len(df_filtered) // 100)
        at_risk_industries = industry_stats[
            (industry_stats['total_articles'] >= min_articles) & 
            (industry_stats['avg_sentiment'] < 0)
        ].sort_values('risk_score', ascending=True)  # Ascending for horizontal bar
        
        if len(at_risk_industries) == 0:
            print("‚ùå No industries with negative sentiment found!")
            return
        
        # Take top 10 most at risk
        at_risk_industries = at_risk_industries.tail(10)
        
        # Create horizontal bar plot
        y_pos = range(len(at_risk_industries))
        colors = plt.cm.Reds(np.linspace(0.4, 0.9, len(at_risk_industries)))
        
        bars = ax.barh(y_pos, at_risk_industries['risk_score'], 
                      color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
        
        ax.set_title('Industries with Most Jobs at Risk\\n(Based on AI Sentiment Analysis)', 
                    fontsize=14, pad=20)
        ax.set_xlabel('Risk Score (Negative Sentiment √ó Risk Proportion √ó log(Articles))', fontsize=12)
        ax.set_ylabel('Industry', fontsize=12)
        
        # Format industry names
        industry_labels = [ind.replace('_', ' ').title() for ind in at_risk_industries.index]
        ax.set_yticks(y_pos)
        ax.set_yticklabels(industry_labels)
        
        # Add value labels
        for i, (bar, idx) in enumerate(zip(bars, at_risk_industries.index)):
            width = bar.get_width()
            row = at_risk_industries.loc[idx]
            ax.text(width + max(at_risk_industries['risk_score']) * 0.01, 
                   bar.get_y() + bar.get_height()/2,
                   f'{row["avg_sentiment"]:.3f}\\n({row["negative_proportion"]:.1%} negative)', 
                   ha='left', va='center', fontsize=8)
        
        ax.grid(True, alpha=0.3, axis='x')
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating industries at risk plot: {e}")
        ax.text(0.5, 0.5, f'Industries at risk plot failed:\\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Industries most at risk plot complete!")

def plot_jobs_at_risk(df, save_path=None):
    """Plot job categories most at risk from AI based on sentiment"""
    print("üìä Creating jobs at risk plot...")
    
    if 'primary_job' not in df.columns:
        print("‚ùå No 'primary_job' column found!")
        return
    
    # Get sentiment column
    sentiment_col = 'sentiment_overall_enhanced' if 'sentiment_overall_enhanced' in df.columns else 'sentiment_base_textblob'
    if sentiment_col not in df.columns:
        print("‚ùå No sentiment column found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    
    try:
        # Filter data
        df_filtered = df.dropna(subset=['primary_job', sentiment_col])
        
        if len(df_filtered) == 0:
            print("‚ùå No valid job and sentiment data!")
            return
        
        # Calculate job risk metrics
        job_stats = df_filtered.groupby('primary_job')[sentiment_col].agg([
            'mean', 'count', 'std',
            lambda x: (x < 0).sum(),  # Negative sentiment count
            lambda x: (x < -0.1).sum(),  # Strongly negative sentiment count
            lambda x: np.percentile(x, 25)  # 25th percentile
        ]).round(3)
        
        job_stats.columns = ['avg_sentiment', 'total_articles', 'sentiment_std', 
                           'negative_count', 'strongly_negative_count', 'q25_sentiment']
        
        # Calculate comprehensive risk score
        job_stats['negative_proportion'] = job_stats['negative_count'] / job_stats['total_articles']
        job_stats['risk_score'] = (
            abs(job_stats['avg_sentiment'].clip(upper=0)) * 
            job_stats['negative_proportion'] * 
            np.log(job_stats['total_articles'] + 1) *
            (1 + abs(job_stats['q25_sentiment'].clip(upper=0)))  # Factor in how consistently negative
        )
        
        # Filter for jobs with sufficient data and negative sentiment
        min_articles = max(20, len(df_filtered) // 200)
        at_risk_jobs = job_stats[
            (job_stats['total_articles'] >= min_articles) & 
            (job_stats['avg_sentiment'] < 0)
        ].sort_values('risk_score', ascending=False)
        
        if len(at_risk_jobs) == 0:
            print("‚ùå No job categories with negative sentiment found!")
            return
        
        # Take top 12 most at risk jobs
        at_risk_jobs = at_risk_jobs.head(12)
        
        # Create bar plot
        x_pos = range(len(at_risk_jobs))
        colors = plt.cm.OrRd(np.linspace(0.5, 0.9, len(at_risk_jobs)))
        
        bars = ax.bar(x_pos, at_risk_jobs['risk_score'], 
                     color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
        
        ax.set_title('Job Categories Most at Risk from AI\\n(Based on Sentiment Analysis)', 
                    fontsize=14, pad=20)
        ax.set_xlabel('Job Category', fontsize=12)
        ax.set_ylabel('Risk Score', fontsize=12)
        
        # Format job names
        job_labels = [job.replace('_', ' ').title() for job in at_risk_jobs.index]
        ax.set_xticks(x_pos)
        ax.set_xticklabels(job_labels, rotation=45, ha='right')
        
        # Add value labels on bars
        for i, (bar, idx) in enumerate(zip(bars, at_risk_jobs.index)):
            height = bar.get_height()
            row = at_risk_jobs.loc[idx]
            ax.text(bar.get_x() + bar.get_width()/2., height + max(at_risk_jobs['risk_score']) * 0.01,
                   f'{row["avg_sentiment"]:.3f}\\n({row["negative_proportion"]:.1%})', 
                   ha='center', va='bottom', fontsize=8)
        
        ax.grid(True, alpha=0.3, axis='y')
        
        # Add risk level annotations
        max_risk = max(at_risk_jobs['risk_score'])
        ax.axhline(max_risk * 0.7, color='red', linestyle='--', alpha=0.5, label='High Risk')
        ax.axhline(max_risk * 0.4, color='orange', linestyle='--', alpha=0.5, label='Medium Risk')
        ax.legend()
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating jobs at risk plot: {e}")
        ax.text(0.5, 0.5, f'Jobs at risk plot failed:\\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Jobs at risk plot complete!")

def plot_jobs_enhanced(df, save_path=None):
    """Plot job categories most enhanced/benefited by AI based on positive sentiment"""
    print("üìä Creating jobs enhanced by AI plot...")
    
    if 'primary_job' not in df.columns:
        print("‚ùå No 'primary_job' column found!")
        return
    
    # Get sentiment column
    sentiment_col = 'sentiment_overall_enhanced' if 'sentiment_overall_enhanced' in df.columns else 'sentiment_base_textblob'
    if sentiment_col not in df.columns:
        print("‚ùå No sentiment column found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    
    try:
        # Filter data
        df_filtered = df.dropna(subset=['primary_job', sentiment_col])
        
        if len(df_filtered) == 0:
            print("‚ùå No valid job and sentiment data!")
            return
        
        # Calculate job enhancement metrics
        job_stats = df_filtered.groupby('primary_job')[sentiment_col].agg([
            'mean', 'count', 'std',
            lambda x: (x > 0).sum(),  # Positive sentiment count
            lambda x: (x > 0.1).sum(),  # Strongly positive sentiment count
            lambda x: np.percentile(x, 75)  # 75th percentile
        ]).round(3)
        
        job_stats.columns = ['avg_sentiment', 'total_articles', 'sentiment_std', 
                           'positive_count', 'strongly_positive_count', 'q75_sentiment']
        
        # Calculate enhancement score
        job_stats['positive_proportion'] = job_stats['positive_count'] / job_stats['total_articles']
        job_stats['enhancement_score'] = (
            job_stats['avg_sentiment'].clip(lower=0) * 
            job_stats['positive_proportion'] * 
            np.log(job_stats['total_articles'] + 1) *
            (1 + job_stats['q75_sentiment'].clip(lower=0))  # Factor in how consistently positive
        )
        
        # Filter for jobs with sufficient data and positive sentiment
        min_articles = max(20, len(df_filtered) // 200)
        enhanced_jobs = job_stats[
            (job_stats['total_articles'] >= min_articles) & 
            (job_stats['avg_sentiment'] > 0)
        ].sort_values('enhancement_score', ascending=False)
        
        if len(enhanced_jobs) == 0:
            print("‚ùå No job categories with positive sentiment found!")
            return
        
        # Take top 12 most enhanced jobs
        enhanced_jobs = enhanced_jobs.head(12)
        
        # Create bar plot
        x_pos = range(len(enhanced_jobs))
        colors = plt.cm.Greens(np.linspace(0.5, 0.9, len(enhanced_jobs)))
        
        bars = ax.bar(x_pos, enhanced_jobs['enhancement_score'], 
                     color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
        
        ax.set_title('Job Categories Most Enhanced by AI\\n(Based on Positive Sentiment Analysis)', 
                    fontsize=14, pad=20)
        ax.set_xlabel('Job Category', fontsize=12)
        ax.set_ylabel('Enhancement Score', fontsize=12)
        
        # Format job names
        job_labels = [job.replace('_', ' ').title() for job in enhanced_jobs.index]
        ax.set_xticks(x_pos)
        ax.set_xticklabels(job_labels, rotation=45, ha='right')
        
        # Add value labels on bars
        for i, (bar, idx) in enumerate(zip(bars, enhanced_jobs.index)):
            height = bar.get_height()
            row = enhanced_jobs.loc[idx]
            ax.text(bar.get_x() + bar.get_width()/2., height + max(enhanced_jobs['enhancement_score']) * 0.01,
                   f'{row["avg_sentiment"]:.3f}\\n({row["positive_proportion"]:.1%})', 
                   ha='center', va='bottom', fontsize=8)
        
        ax.grid(True, alpha=0.3, axis='y')
        
        # Add enhancement level annotations
        max_enhancement = max(enhanced_jobs['enhancement_score'])
        ax.axhline(max_enhancement * 0.7, color='green', linestyle='--', alpha=0.5, label='High Enhancement')
        ax.axhline(max_enhancement * 0.4, color='lightgreen', linestyle='--', alpha=0.5, label='Medium Enhancement')
        ax.legend()
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating jobs enhanced plot: {e}")
        ax.text(0.5, 0.5, f'Jobs enhanced plot failed:\\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ Jobs enhanced plot complete!")

def plot_ai_technology_mention_distribution(df, save_path=None):
    """Plot distribution of actual AI technology mentions from the data"""
    print("üìä Creating AI technology mention distribution...")
    
    if 'ai_technologies' not in df.columns:
        print("‚ùå No 'ai_technologies' column found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(14, 10))
    ax = fig.add_subplot(111)
    
    try:
        # Extract actual technology mentions from the data
        tech_counts = defaultdict(int)
        
        for tech_dict in df['ai_technologies'].dropna():
            if isinstance(tech_dict, dict):
                for tech_category, tech_list in tech_dict.items():
                    if isinstance(tech_list, list):
                        for tech in tech_list:
                            tech_counts[tech] += 1
                    else:
                        tech_counts[tech_category] += 1
        
        if not tech_counts:
            print("‚ùå No AI technology mentions found in data!")
            return
        
        # Sort by count (descending) and take top 20
        sorted_techs = sorted(tech_counts.items(), key=lambda x: x[1], reverse=True)[:20]
        
        if not sorted_techs:
            print("‚ùå No technology data to display!")
            return
        
        technologies, counts = zip(*sorted_techs)
        
        # Create color gradient
        colors = plt.cm.viridis(np.linspace(0.2, 0.9, len(technologies)))
        
        # Create horizontal bar plot
        y_pos = range(len(technologies))
        bars = ax.barh(y_pos, counts, color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
        
        # Customize plot
        ax.set_title('AI Technology Mention Distribution', fontsize=16, pad=20)
        ax.set_xlabel('Number of Mentions', fontsize=12)
        ax.set_ylabel('Technology', fontsize=12)
        ax.set_yticks(y_pos)
        
        # Clean up technology names for display
        clean_tech_names = [tech.replace('_', ' ').title() for tech in technologies]
        ax.set_yticklabels(clean_tech_names)
        
        # Add value labels on bars
        for i, (bar, count) in enumerate(zip(bars, counts)):
            width = bar.get_width()
            ax.text(width + max(counts) * 0.01, bar.get_y() + bar.get_height()/2,
                   f'{count:,}', ha='left', va='center', fontsize=9, fontweight='bold')
        
        # Invert y-axis to show highest counts at top
        ax.invert_yaxis()
        ax.grid(True, alpha=0.3, axis='x')
        
        # Add statistics
        total_mentions = sum(counts)
        unique_technologies = len(tech_counts)
        ax.text(0.98, 0.02, f'Total Mentions: {total_mentions:,}\\nUnique Technologies: {unique_technologies}', 
                transform=ax.transAxes, ha='right', va='bottom',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
    except Exception as e:
        print(f"‚ö†Ô∏è Error creating technology distribution plot: {e}")
        ax.text(0.5, 0.5, f'Technology distribution plot failed:\\n{str(e)}', 
                ha='center', va='center', transform=ax.transAxes)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"üíæ Plot saved to {save_path}")
    
    plt.show()
    plt.close(fig)
    print("‚úÖ AI technology mention distribution plot complete!")

def create_extended_sentiment_dashboard(df):
    """Create comprehensive dashboard with all plots including the new ones"""
    print("üéØ Creating EXTENDED comprehensive sentiment analysis dashboard...")
    print("=" * 80)
    
    # Setup plotting environment first
    setup_plotting_environment()
    
    # Test plotting
    print("üß™ Testing plotting configuration...")
    if not test_plotting():
        print("‚ùå Plotting test failed! Please check your environment.")
        return False
    
    print(f"üìä Dataset info: {df.shape[0]:,} articles, {df.shape[1]} columns")
    
    # All plots including the new ones
    plots = [
        ("Sentiment Distribution", plot_sentiment_distribution),
        ("Sentiment by Industry", plot_industry_sentiment), 
        ("Sentiment by Job Category", plot_job_sentiment),
        ("AI Technology Mentions", plot_technology_mentions),
        ("Sentiment Methods Comparison", plot_sentiment_comparison),
        ("AI Technology Adoption Over Time", plot_ai_technology_adoption_over_time),
        ("Topic Correlation Heatmap", plot_topic_correlation_heatmap),
        ("AI-Job Categories Network", plot_ai_job_network),
        ("AI Technology Mention Distribution", plot_ai_technology_mention_distribution)
    ]
    
    success_count = 0
    for i, (title, plot_func) in enumerate(plots, 1):
        print(f"\\n{i}. {title}:")
        print("-" * 50)
        try:
            plot_func(df)
            success_count += 1
        except Exception as e:
            print(f"‚ùå Error in {title}: {e}")
            import traceback
            traceback.print_exc()
    
    print(f"\\nüéâ Extended dashboard complete! {success_count}/{len(plots)} plots successful")
    return success_count == len(plots)

# Load from cache function (if not already defined)
def load_from_cache(filename):
    """Load object from cache if it exists"""
    import pickle
    import os
    
    cache_dir = "cache"
    cache_path = os.path.join(cache_dir, filename)
    
    if os.path.exists(cache_path):
        try:
            with open(cache_path, 'rb') as f:
                return pickle.load(f)
        except Exception as e:
            print(f"‚ö†Ô∏è Error loading {filename}: {e}")
            return None
    else:
        print(f"‚ö†Ô∏è Cache file {cache_path} not found")
        return None
```

```python
#AI sentiment timeline plot with key events
def plot_ai_sentiment_timeline(df, save_path=None):
    """Plot AI sentiment timeline with key events"""
    print("Creating AI sentiment timeline plot...")
    
    if 'date' not in df.columns or 'sentiment_base_textblob' not in df.columns:
        print("Required columns 'date' and 'sentiment_base_textblob' not found!")
        return
    
    # Create figure
    fig = plt.figure(figsize=(14, 8))
    ax = fig.add_subplot(111)
    
    try:
        # Filter data
        df_filtered = df.dropna(subset=['date', 'sentiment_base_textblob'])
        
        if len(df_filtered) == 0:
            print("No valid date and sentiment data!")
            return
        
        # Convert date to datetime
        df_filtered['date'] = pd.to_datetime(df_filtered['date'])
        
        # Calculate daily average sentiment
        daily_sentiment = df_filtered.groupby(df_filtered['date'].dt.date)['sentiment_base_textblob'].mean().reset_index()
        
        if daily_sentiment.empty:
            print("No data available for plotting!")
            return
        
        # Plot the sentiment over time
        ax.plot(daily_sentiment['date'], daily_sentiment['sentiment_base_textblob'], 
                color='blue', linewidth=2, label='Daily Average Sentiment')
        
        # Add key events (example events)
        key_events = {
            '2023-01-01': 'Event A',
            '2023-02-15': 'Event B',
            '2023-03-10': 'Event C'
        }
        
        for event_date, event_label in key_events.items():
            event_date_dt = pd.to_datetime(event_date)
            if event_date_dt in daily_sentiment['date'].values:
                ax.axvline(x=event_date_dt, color='red', linestyle='--', alpha=0.5)
                ax.text(event_date_dt, 0.5, event_label, rotation=45, ha='left', va='center', 
                        fontsize=10, color='red')
        
        # Customize plot
        ax.set_title('AI Sentiment Timeline with Key Events', fontsize=16, pad=20)
        ax.set_xlabel('Date', fontsize=12)
        ax.set_ylabel('Average Sentiment Score', fontsize=12)
        
        # Adjust layout
        plt.tight_layout()
            
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Plot saved to {save_path}")
            plt.show()
            plt.close(fig)
            print("AI sentiment timeline plot complete!")
        except Exception as e:
            print(f"‚ö†Ô∏è Error creating plot: {e}")
            plt.tight_layout()
            plt.show()
            plt.close(fig)
            print("Plot failed!")
```
